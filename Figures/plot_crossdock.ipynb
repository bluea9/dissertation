Plotting Separated interface energy vs delta SASA by antibody

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np

complex_list = ['3MXW', '2VXT','3EO1', '3HI6','3G6D']
results_dir = os.path.expanduser('~/MSC/Results/Top100/Energy_scores_cross_passive')
energy_scores_file = os.path.expanduser('~/MSC/Results/Top100/top_energy_scores_passive.csv')
df = pd.read_csv(energy_scores_file)
energy_scores_real = os.path.expanduser('~/MSC/Results/Top100/top_energy_scores_real.csv')
df_real = pd.read_csv(energy_scores_real)
energy_scores_af = os.path.expanduser('~/MSC/Results/Top100/top_energy_scores_real_af.csv')
df_af = pd.read_csv(energy_scores_af)

colors_list = ['red', 'blue', 'green', 'orange', 'purple']

df['Color'] = 'nb'

for id in complex_list:
    plt.figure(figsize=(10, 6))
    for id2, color in zip(complex_list, colors_list): 
        if id != id2:
            combination = id2 + '_' + id
            df_id2 = df[df['Complex'].str.contains(combination, na=False)]
            df_id2['Color'] = color
            colors = np.array(df_id2['Color'])
            plt.scatter(df_id2['Interface_delta_SASA'], df_id2['Separated_interface_energy'], marker='.', c=colors, label=id2+' - '+id, alpha=0.8, s=100)
        else: 
            real_pair = id+'_'+id
            df_id = df[df['Complex'].str.contains(real_pair, na=False)]
            df_id.loc[df['Complex'].str.contains(real_pair, na=False), 'Color'] = 'black'
            colors = np.array(df_id['Color'])
            plt.scatter(df_id['Interface_delta_SASA'], df_id['Separated_interface_energy'], marker='.', c=colors, label=id+' - '+id, alpha=0.8, s=100)      
    df_af2 = df_af[df_af['Complex'].str.contains(id, na=False)]
    df_af2['Color'] = 'black'
    colors = np.array(df_af2['Color'])
    plt.scatter(df_af2['Interface_delta_SASA'], df_af2['Separated_interface_energy'], marker='D', c=colors, label=id + ' AF', alpha=0.8, s=15)
    df_real2 = df_real[df_real['Complex'].str.contains(id, na=False)]
    df_real2.loc[df_real['Complex'].str.contains(id, na=False), 'Color'] = 'black'
    colors = np.array(df_real2['Color'])
    plt.scatter(df_real2['Interface_delta_SASA'], df_real2['Separated_interface_energy'], marker='*', c=colors, label=id + ' Real', alpha=0.8, s=50)
    plt.title(f'Antigen from complex {id}', fontweight='bold')
    plt.xlabel('Interface ΔSASA (Å²)')
    plt.ylabel('Separated interface energy (REU)')
    plt.legend(title='Combinations', loc='upper right')
    plt.grid(True)
    graph_file = os.path.join(results_dir, id + '_ag_energy_sc_cross_passive.png')
    plt.savefig(graph_file, format='png', dpi=300)
    plt.show()
```

Violin plots of ΔSASA of crossdocking experiments

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
import seaborn as sns

complex_list = ['3MXW', '2VXT','3EO1', '3HI6','3G6D']
results_dir = os.path.expanduser('~/MSC/Results/Top100/Violin_plots')
energy_scores_file = os.path.expanduser('~/MSC/Results/Top100/top_energy_scores_passive.csv')
df = pd.read_csv(energy_scores_file)
energy_scores_real = os.path.expanduser('~/MSC/Results/Top100/top_energy_scores_real.csv')
df_real = pd.read_csv(energy_scores_real)
energy_scores_af = os.path.expanduser('~/MSC/Results/Top100/top_energy_scores_real_af.csv')
df_af = pd.read_csv(energy_scores_af)
energy_scores_hv = os.path.expanduser('~/MSC/Results/Top100/top_energy_scores_hv_epi.csv')
df_hv = pd.read_csv(energy_scores_hv)

for id in complex_list:
    df_real2 = df_real[df_real['Complex'].str.startswith(id, na=False)]
    df_real3 = df_real2[['Interface_delta_SASA']]
    df_real3 = df_real3.rename(columns={'Interface_delta_SASA': f'{id} - Real'})
    df_real3 = df_real3.reset_index(drop=True)

    df_af2 = df_af[df_af['Complex'].str.startswith(id, na=False)]
    df_af3 = df_af2[['Interface_delta_SASA']]
    df_af3 = df_af3.rename(columns={'Interface_delta_SASA': f'{id} - AF'})
    df_af3 = df_af3.reset_index(drop=True)

    df_hv2 = df_hv[df_hv['Complex'].str.startswith(id, na=False)]
    df_hv3 = df_hv2[['Interface_delta_SASA']]
    df_hv3 = df_hv3.rename(columns={'Interface_delta_SASA': f'{id} - HV Epi'})
    df_hv3 = df_hv3.reset_index(drop=True)

    df_partial = pd.concat([df_real3, df_af3, df_hv3], axis=1)
        
    df_sasa = df_partial
    for id2 in complex_list:
        df_2 = df[df['Complex'].str.startswith(id+'_'+id2, na=False)]
        df_3 = df_2[['Interface_delta_SASA']]
        df_3 = df_3.rename(columns={'Interface_delta_SASA': f'{id} - {id2}'})
        df_3 = df_3.reset_index(drop=True)
        df_sasa = pd.concat([df_sasa, df_3], axis=1)
    filename = os.path.join(results_dir, id + '_SASA_passive.csv')
    df_sasa.to_csv(filename, index=False)
    sns.violinplot(data=df_sasa, palette='colorblind')
    plt.xlabel('Docking scenario')
    plt.ylabel('Interface ΔSASA (Å²)')
    plt.xticks(rotation=90)
    plt.tight_layout()
    graph_file = os.path.join(results_dir, id + '_SASA_passive.png')
    plt.savefig(graph_file, format='png', dpi=300)
    plt.show()
```
